```python
# Importación de librerías del entorno científico
import pandas as pd
import numpy as np

# Configuración de reproducibilidad técnica (Semilla global)
# Garantiza que cualquier partición o muestreo futuro genere resultados idénticos
RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)

print(f"Librerías importadas correctamente. Semilla de aleatoriedad fijada en: {RANDOM_STATE}")
```

## 3. Adquisición y Carga de Datos
Dado que el dataset original (`cybersecurity_threat_detection_logs.csv`) cuenta con aproximadamente 6 millones de registros, la lectura directa puede generar advertencias de desbordamiento de memoria por tipos de datos mixtos (*mixed dtypes*). Para mitigar esto y asegurar una ejecución estable, la carga se realiza mediante una función modular utilizando el parámetro `low_memory=False` de Pandas.

```python
def cargar_logs_seguridad(ruta_archivo: str) -> pd.DataFrame:
    """
    Lee el dataset masivo de logs y devuelve un DataFrame.
    Se utiliza low_memory=False para asegurar la integridad de la lectura en datasets de gran volumen.
    """
    try:
        df = pd.read_csv(ruta_archivo, low_memory=False)
        print("Dataset cargado exitosamente.")
        return df
    except FileNotFoundError:
        print(f"Error: No se encontró el archivo en la ruta {ruta_archivo}.")
        return None

def explorar_dimensiones(df: pd.DataFrame) -> None:
    """Imprime el tamaño y los tipos de datos principales del DataFrame"""
    if df is not None:
        filas, columnas = df.shape
        print(f"\n--- Exploración Inicial ---")
        print(f"Total de logs: {filas:,}")
        print(f"Total de columnas: {columnas}")
        print("\nTipos de datos detectados:")
        print(df.dtypes)

# Ejecución del flujo reproducible apuntando a la carpeta de datos originales
# IMPORTANTE guardar los datos originales en este directorio
ruta_dataset = '../data/raw/cybersecurity_threat_detection_logs.csv'

df_amenazas = cargar_logs_seguridad(ruta_dataset)
explorar_dimensiones(df_amenazas)
```
