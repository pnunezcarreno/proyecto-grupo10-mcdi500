## 1. Contexto y Problema Algorítmico
Para optimizar el triaje en el SOC, necesitamos ordenar rápidamente los eventos de red según su volumen de transferencia (`bytes_escalados`). Esto permite priorizar la investigación de posibles exfiltraciones masivas de datos.

> **Referencia al Foro Técnico (Semana 1):** Tal como se discutió en el foro de la semana 1, la calidad de los datos y el preprocesamiento exhaustivo (eliminación de nulos, normalización y transformación de formatos) son críticos para cualquier pipeline analítico. El éxito y la precisión de los algoritmos de ordenamiento que evaluaremos a continuación dependen de manera absoluta de esa matriz limpia y numéricamente estandarizada que construimos en la Fase 2; sin ese preprocesamiento previo, las operaciones algorítmicas generarían errores de ejecución o resultados inválidos.

```python
import pandas as pd
import timeit
import sys
import os

# Aumentamos el límite de recursión por seguridad en algoritmos profundos
sys.setrecursionlimit(5000)

def cargar_datos_procesados(ruta):
    """
    Carga el dataset masivo limpio de la Fase 2 con validación defensiva.
    """
    # Validación defensiva por si se clonó de GitHub sin correr la Fase 2
    if not os.path.exists(ruta):
        raise FileNotFoundError(
            f"\n[ERROR DE ENTORNO] No se encontró el archivo: {ruta}\n"
            f"Como el dataset masivo se excluye de GitHub por peso, "
            f"se debe ejecutar primero el notebook de la Fase 2 'F2_Preprocesamiento.ipynb' "
            f"para generar este archivo localmente antes de evaluar la Fase 3."
        )
        
    print(f"Cargando el dataset masivo desde: {ruta}...")
    
    # Transformación mediante selección estricta: Solo cargamos las columnas que necesitamos
    columnas_necesarias = ['timestamp', 'bytes_escalados', 'is_automated_ua', 'threat_label_encoded']
    df = pd.read_csv(ruta, usecols=columnas_necesarias)
    
    print(f"¡Carga exitosa! Datos cargados: {df.shape[0]:,} filas.")
    return df

ruta_procesados = '../data/processed/cybersecurity_logs_procesados.csv'
df_f3 = cargar_datos_procesados(ruta_procesados)
display(df_f3.head(3))
```

## 2. Preprocesamiento Mínimo y Transformación Estructural
Dado que la Fase 2 garantizó una matriz sin nulos y escalada numéricamente, omitiremos rutinas de limpieza adicionales. 

La única transformación implementada fue la **conversión estructural** de la columna objetivo (`bytes_escalados`) a una lista nativa de Python (`List`).
* **Justificación técnica:** Los algoritmos iteran mucho más rápido sobre listas simples. Hacerlo directamente sobre un *DataFrame* añade latencia (*overhead*) que distorsiona la medición real de la complejidad algorítmica.

```python
print("--- INICIANDO TRANSFORMACIÓN ESTRUCTURAL ---")

# Mostramos el tipo de dato original (Pandas Series)
tipo_original = type(df_f3['bytes_escalados'])
print(f"Tipo de dato original: {tipo_original}")

# Aplicamos la transformación estructural explícita (.tolist)
lista_bytes_nativa = df_f3['bytes_escalados'].tolist()

# Mostramos el tipo de dato transformado (Python List)
tipo_transformado = type(lista_bytes_nativa)
print(f"Tipo de dato transformado: {tipo_transformado}")
print(f"Total de elementos en la lista: {len(lista_bytes_nativa):,}")
print("Transformación estructural completada. Datos listos para inyección algorítmica.")
```
